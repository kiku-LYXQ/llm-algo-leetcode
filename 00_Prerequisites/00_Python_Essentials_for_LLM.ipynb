## Part 1: Python 基础与数据结构

### 练习 1.1: 生成位置索引
实现一个函数，返回长度为 `seq_len` 的位置索引列表。


```python
def generate_position_ids(seq_len):
    """返回 [0, 1, ..., seq_len-1]。"""
    return [idx for idx in range(seq_len)]


def merge_model_config(base_config, overrides):
    """合并模型配置，override 覆盖 base 中的同名键。"""
    merged = base_config.copy()
    merged.update(overrides)
    return merged


def count_token_frequency(tokens):
    """统计 token 频率。"""
    freq = {}
    for token in tokens:
        freq[token] = freq.get(token, 0) + 1
    return freq


class ModelSpec:
    def __init__(self, name, num_params_b, precision='bf16'):
        self.name = name
        self.num_params_b = num_params_b
        self.precision = precision

    def summary(self):
        return f"{self.name} ({self.num_params_b:.1f}B, {self.precision})"
```

```python
def test_generate_position_ids():
    assert generate_position_ids(5) == [0, 1, 2, 3, 4]
    assert generate_position_ids(1) == [0]
    print('✅ generate_position_ids 通过')


def test_merge_model_config():
    base = {'hidden_size': 1024, 'num_layers': 24, 'precision': 'bf16'}
    overrides = {'num_layers': 32, 'dropout': 0.1}
    merged = merge_model_config(base, overrides)
    assert merged == {'hidden_size': 1024, 'num_layers': 32, 'precision': 'bf16', 'dropout': 0.1}
    assert base['num_layers'] == 24
    print('✅ merge_model_config 通过')


def test_count_token_frequency():
    tokens = ['<bos>', 'hello', 'hello', 'world', 'world', 'world']
    assert count_token_frequency(tokens) == {'<bos>': 1, 'hello': 2, 'world': 3}
    print('✅ count_token_frequency 通过')


def test_model_spec():
    spec = ModelSpec('TinyLM', 1.5, 'fp16')
    assert spec.summary() == 'TinyLM (1.5B, fp16)'
    print('✅ ModelSpec 通过')


test_generate_position_ids()
test_merge_model_config()
test_count_token_frequency()
test_model_spec()
```

## Part 2: 实战应用

### 练习 2.1: 配置一个小模型
把一个基础配置和实验覆盖项合并，模拟训练脚本中的参数读取。


```python
base_config = {'model_name': 'TinyLM', 'hidden_size': 768, 'num_layers': 12, 'precision': 'bf16'}
overrides = {'precision': 'fp16', 'batch_size': 8}
config = merge_model_config(base_config, overrides)

print('合并后的配置：')
for key, value in config.items():
    print(f'{key}: {value}')

print()
print('位置索引示例：', generate_position_ids(8))
print('token 统计示例：', count_token_frequency(['llm', 'llm', 'algo']))
```
